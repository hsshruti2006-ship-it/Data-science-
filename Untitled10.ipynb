```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame with 'Age' and 'Salary' columns
data = {
    'Age': np.random.randint(20, 60, 10),
    'Salary': np.random.randint(30000, 100000, 10)
}
df = pd.DataFrame(data)

# Sort by 'Age' for a better-looking area plot
df = df.sort_values(by='Age').reset_index(drop=True)

plt.fill_between(df['Age'], df['Salary'], color='skyblue', alpha=0.4)
plt.plot(df['Age'], df['Salary'], color='Slateblue', alpha=0.6) # Optional: add a line on top
plt.title('Area Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.stem(df['Age'],df['Salary'])
plt.title('Stem Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.stackplot(df['Age'],df['Salary'])
plt.title('Stack Plot')
plt.show()
```
